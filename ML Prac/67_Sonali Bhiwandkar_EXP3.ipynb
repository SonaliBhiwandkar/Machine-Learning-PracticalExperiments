```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Given data
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1.2, 1.8, 2.6, 3.2, 3.8])

# Convert to DataFrame
data = pd.DataFrame({'x': x.flatten(), 'y': y})
print(data.shape)
print(data.head())
```

```python
# Step 1: Calculate means
x_mean = data['x'].mean()
y_mean = data['y'].mean()
```

```python
# Step 2: Calculate covariance and variance of x
covariance = np.sum((data['x'] - x_mean) * (data['y'] - y_mean))
variance = np.sum((data['x'] - x_mean) ** 2)
```

```python
# Step 3: Calculate slope (m) and y-intercept (b)
m = covariance / variance
b = y_mean - m * x_mean
```

```python
# Step 4: Predict values of y
data['y_pred'] = m * data['x'] + b
```

```python
# Step 5: Plot data and regression line
plt.scatter(data['x'], data['y'], label='Data points')
plt.plot(data['x'], data['y_pred'], color='red', label='Regression line')
plt.title('Simple Linear Regression (Least Squares Method)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate R-squared (accuracy)
total_variation = np.sum((data['y'] - y_mean) ** 2)
residual_variation = np.sum((data['y'] - data['y_pred']) ** 2)
r_squared = 1 - (residual_variation / total_variation)
```

```python
# Print the equation of the line and accuracy
print("Regression Equation: y = {:.2f} + {:.2f} * x".format(b, m))
print("Accuracy (R-squared): {:.2f}".format(r_squared))
```
