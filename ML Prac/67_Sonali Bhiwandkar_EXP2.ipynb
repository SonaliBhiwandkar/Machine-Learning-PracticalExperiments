```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5]).reshape(-1, 1)
y = np.array([1.2, 1.8, 2.6, 3.2, 3.8])

model = LinearRegression().fit(x, y)

# Predicted value for x=7
predicted_y = model.predict(np.array([[7]]))

# Print regression equation coefficients
slope = model.coef_[0]
print("Slope is : ",slope)
intercept = model.intercept_
print("Regression Equation: y = {:.2f} + {:.2f} * x".format(intercept, slope))

# Print predicted value for x=7
print("Predicted value of y for x=7 : ", predicted_y[0])

# Calculate R-squared
y_pred = model.predict(x)
accuracy = r2_score(y, y_pred)
print("Accuracy of the model: {:.2f}%".format(accuracy * 100))

# Plot
plt.scatter(x, y, label='Data points')
plt.plot(x, model.predict(x), color='red', label='Regression line')
plt.title('Simple Linear Regression')
plt.xlabel('x')
plt.ylabel('y')

# Adding slope and accuracy to the plot
plt.text(3, 2.5, "Slope = {:.2f}".format(slope), fontsize=12, color='green')
plt.text(3, 2, "Accuracy = {:.2f}%".format(accuracy * 100), fontsize=12, color='blue')

plt.legend()
plt.grid(True)
plt.show()
```
